```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
#convertir data frame
df = pd.read_csv("Datos_Facturación.csv")
df
```

```python
# 1) CVE_CLPV entre 1000 y 2000 (incluye límites)
filtro1 = df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
print(filtro1.head())
```

```python
filtro2 = df[(df["CVE_VEND"] != 4) & (df["CVE_VEND"] != 5)]
print(filtro2.head())
```

```python
fecha_objetivo = pd.to_datetime("28/02/2022", dayfirst=True)
filtro3 = df[df["FECHA_ENT"] == fecha_objetivo]

print(filtro3.head())
```

```python
filtro4 = df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "*E")]
print(filtro4.head())
```

```python
filtro5 = df.loc[:, ["CVE_DOC", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]
print(filtro5.head())
```

```python
filtro6 = df.iloc[7001:7100]   # 7099 incluido porque Python excluye el límite superior
print(filtro6.head())
```

```python
filtro7 = df[df["CVE_VEND"].isin([1, 2])]
print(filtro7.head())
```
